# **INTRODUCTION**

Les complexes de collecte de lumière (Light Harvesting Complexes, LHCs) sont un ensemble de protéines et de pigments photosynthétiques organisés dans les cellules des organismes capables de réaliser la photosynthèse, comme les plantes, les algues et certaines bactéries. Il a pour rôle principal de capter l'énergie lumineuse du soleil et de la transférer avec une efficacité quantique (Quantum Efficiency, QE) de près de 100% vers les centres réactionnels de la photosynthèse. Le flux de photons dans lequel opèrent les organismes photosynthétiques est faible, ce qui signifie que la probabilité d'absorber un photon est faible. Il est donc d'une importance cruciale que le processus de captation de la lumière qui suit soit très efficace. C'est particulièrement vrai pour les bactéries vertes du soufre Prosthecochloris aestuarii et Chlorobium tepidum, adaptées aux faibles éclairages. L'énergie de toute lumière solaire qu'elles absorbent est transférée entre trois "corps" de chlorophylle. Le premier, celui principalement excité par la lumière solaire, est connu sous le nom de chlorosome (ou antenne photosynthétique) et contient environ 95% de toute la chlorophylle de ces organismes. Ceux-ci se trouvent sous la membrane cytoplasmique de la cellule. L'énergie d'excitation est transférée via le deuxième corps de chlorophylle, un complexe pigment-protéine de bactériochlorophylle-a (ou BChl), jusqu'au troisième, une paire de molécules de chlorophylle connues sous le nom de paire spéciale du centre réactionnel. C'est là que se produit une séparation de charge et que l'énergie de photo-excitation est convertie en énergie chimique sous forme de gradient électrochimique transmembranaire (voir figure ci-dessous). Comprendre les principes qui régissent cette efficacité pourrait inspirer le développement de nouvelles technologies énergétiques plus performantes, et alors permettre le développement de cellules solaires organiques et hybrides s'inspirant des principes de collecte et de transfert de l'énergie lumineuse du complexe naturel, et la Conception de systèmes photovoltaïques plus efficaces en s'appuyant sur les mécanismes quantiques observés dans le complexe. Toute fois, des études ont montré que les effets quantiques tels que la cohérence quantique, joueraient un rôle très important dans les processus de collecte de lumière photosynthétique. L'étude en profondeur d'un tel système s'avérait alors être très importante.

## complexe FMO

Le complexe de collecte de lumière est un ensemble de plusieurs protéines parmi lesquelles on distingue le complexe FMO qui est une protéine d'échafaudage qui fait le lien entre l'antenne collectrice de lumière et le centre réactionnel photosynthétique. Ce complexe offre un modèle d'étude de choix pour comprendre les mécanismes fondamentaux du transfert d'énergie dans les systèmes photosynthetiques. Des recherches récentes ont notamment montré que le complexe FMO exploite des phénomènes quantiques comme l'intrication et la cohérence quantique pour optimiser le transport de l'énergie. Cela en fait un sujet fascinant à l'intersection de la biologie, de la physique quantique.

Le complexe Fenna-Matthews-Olson (FMO) est un complexe de protéines hydrosolubles assurant le transfert d'énergie d'excitation chez les bactéries sulfureuses vertes telles que Chlorobium Tepidum et Prosthecochloris aestuarii. La protéine  tire son nom de Roger Fenna et Brian Matthews, qui ont été les premiers à déterminer sa structure et de John Olson, qui a découvert la protéine. Sa structure est celle d'un trimère, dans lequel chacune des sous-unités est composée de 8 molécules de bactériochlorophylle-a ({Bchl}-a) ou sites. Le huitième site a été découvert récemment mais son rôle est négligeable. La nomenclature établie pour la numérotation des pigments est celle initialement proposée par Fenna et Matthews, à laquelle a été ajoutée la huitième BChl. Selon cette nomenclature, BChl 3 et BChl 4 sont les plus proches de la membrane cytoplasmique renfermant le centre de la reaction, tandis que BChl 1, BChl 6 et BChl 8 sont plus proches de la plaque de base des chlorosomes qui renferme les pigments d'antenne captant la lumière.

## Modelisation quantique

En raison des observations expérimentales et du fait de la vitesse élevée à laquelle se produit le EET, juste une excitation unique est considérée dans le systeme. Chaque BChl se retrouve donc dans deux états d'excitation : l'état fondamental $|0\rangle$  et le premier état excité $|1\rangle$. Ce qui signifie que lorsqu'un site est excitée, tous les autres sont à l'état fondamental. Notre système peut donc se restreindre à un système à deux niveaux pour chaque BChl. Le premier état excité de chaque molécule Bchl est décrit comme une ***quasi-particule*** d'exciton, c'est-à-dire une paire électron-trou couplé par une interaction Coulombienne. Les énergies de chaque site et celles des couplages entre sites sont résumées dans l'Hamiltonien de Frenkel 

\begin{equation}
      H_S = \sum_{i=1}^{n} |i\rangle \langle i| E_i + \sum_{i\ne j=1}^{n} |i\rangle \langle j| V_{ij}.
\end{equation}


 $E_i$ est l'énergie du site $i$ quand tous les autres sont à l'état fondamental, $n$ le nombre de sites et $V_{ij}$ les forces de couplage Coulombiennes ou d'interaction entre les sites $i$ et $j$. Physiquement, ces forces représentent le recouvrement des dipôles de transition électronique de site, sous l'hypothèse qu'ils sont suffisamment espacés pour qu'aucun échange ne se produise.

 L'environnement ou bain de phonons protéique dans lequel baigne le complexe est un ensemble d'oscillateurs harmoniques de modes k, vibrant à la fréquence $\omega_k$. Ce bain induit des sauts quantiques, la décohérence et le déphasage des excitons sans changer le nombre d'excitations. L'hamiltonien de couplage des phonons s'écrit sous la forme

 \begin{equation}
     H_B = \sum_{i=1}^{n} \sum_{k=1} \frac{\hbar \omega _{k,i}}{2} \Big(  \mathbf{P}^{2}_{k,i} +  \mathbf{Q}^2_{k,i} \Big)
 \end{equation}

 et l'amortissement de l'excitation due au champ de rayonnement est donné par l'Hamiltonien (Hamiltonien d'interaction phonon-site)

 \begin{equation}
     H_{SB} =  \sum_{i=1}^{n} \sum_{k=1} |i\rangle \langle i| g_{k,i} \mathbf{Q}_{k,i}.
 \end{equation}


$ \mathbf{P}_{k,i}$, $\mathbf{Q}_{k,i}$, tous deux sans dimension, sont respectivement l'impulsion et la coordonnée, et $\omega _{k,i}$ la fréquence de vibration, du mode $k$ associé au site $i$. $c_{k,i}$ est la force de couplage entre le site $i$ et le mode $k$ de son environnement proté\"ique. L'opérateur de couplage phonon-site est $     |i\rangle \langle i|$. Ce couplage entre (S) et (B) est supposé être linéaire suivant le déplacement $\mathbf{Q}_{k,i}$ de l'oscillateur. $H_{SB}$ représente les fluctuations sur les sites d'énergies causées par la dynamique des oscillateurs. Il facilite le flux d'énergie vers le RC et induit le déphasage pour le $i^{\text{ieme}}$ site lorsqu'il est à l'état excité.
\\
Ainsi le hamiltonien resultant s'ecrit donc sous la forme
\begin{equation*}
    H = H_S + H_B + H_{SB}
\end{equation*}

En definissant l'operateur statistique comme etant une grandeur mathematique utiliser pour decrire notre systeme, il sera alors definit comme etant un ***produit tensoriel*** entre l'operateur statistique du systeme et celui de l'environnement: on aura alors

\begin{equation*}
    \rho = \rho_S \otimes \rho_B
\end{equation*}

L'etat general d'un exciton peut etre exprimer par
\begin{equation*}
    |a\rangle = \sum_i^N c_i(a)|i\rangle 
\end{equation*}

avec N le nombre de sites (7 pour le FMO) et la norme au carré de chaque coefficients $c_i$ represente la probabilité que le $a^{ieme}$ exciton soit localisé sur le $i^{eme}$ site

Pour l'occupation d'une seule excitation dans le complexe, la somme des probabilités de chaque exciton doit être égale à 1. on a 
$\sum_i^N|c_i|^2=1$


On definit alors un rapport de participation inverse (IPR) peut être calculé pour chaque exciton. Il est défini, pour l'exciton a, comme :

\begin{equation*}
    IPR(a) = \Big( \sum_i^N(c_i(a))^4 \Big)^{-1}
\end{equation*}

# Processus dynamique

Les processus qui se produisent dans le complexe FMO dans le cadre du transfert d'énergie excitonique (EET) peuvent être décrits par l'action d'un superopérateur sur l'operatuer statistique du système réduit. Cela donne une équation générale du mouvement connue sous le nom d'équation de Liouville, qui peut être utilisée comme point de départ généralisé pour décrire la dynamique de l'EET dans le complexe FMO, sous la forme :

 \begin{equation}
     \frac{d}{dt}\rho(t) = L \rho(t)
 \end{equation}

Un superopérateur est un outil mathématique utilisé pour décrire l'évolution d'un système quantique ouvert, c'est-à-dire un système interagissant avec son environnement.

* Les superopérateurs décrivent comment un état quantique évolue lorsque le système interagit avec son environnement.
* Ils permettent de prendre en compte les effets de la décohérence et de la dissipation dans l'évolution du système.

* Un superopérateur est représenté par une matrice d'opérateurs agissant sur l'espace des états du système.
* Chaque élément de la matrice est lui-même un opérateur quantique.

* Les superopérateurs sont des applications complètement positives et préservent la trace (la norme) de l'état du système.
* Ils décrivent une évolution non unitaire du système, contrairement à l'évolution unitaire décrite par l'équation de Schrödinger pour un système fermé.


$L$ est le superoperateur de Liouville qui est exprimé comme la somme des differents superoperateurs individuels

 \begin{equation}
     L = L_{sys} + L_{bath} + L_{trap} + L_{decay} + L_{rad}
 \end{equation}



Chacun de ces termes capture un processus dynamique différent qui se produit lors du transfert de l'EET.

on a donc 



* $L_{sys}$  qui représente le commutateur avec l'hamiltonien du système, qui décrit l'évolution unitaire gouvernée par Hsys tel que

 \begin{equation}
     L_{sys} = -\frac{i}{\hbar} \Big (H_{sys} \otimes \mathbb{I} - \mathbb{I} \otimes H_{sys}^{\dagger} \Big )
 \end{equation}

 avec $\mathbb{I}$ la matrice identité et $\dagger$ represente le conjugué complexe du Hamiltonien du systeme



* $L_{bath}$ qui décrit le processus de relaxation du système après l'excitation photonique, capturant l'influence du bain sur celui-ci sans        définir explicitement la forme du bain. Afin de rendre compte de l'influence du bain purement en termes des degrés de liberté du système, le       superopérateur de relaxation du bain Lbath aura une forme fonctionnelle qui est dérivée en faisant des hypothèses sur la façon dont le système et  le bain interagissent. 

  \begin{equation}
     L_{bath} =  \sum_{\alpha} k_{\alpha} \Big (A_{\alpha}^{*} \otimes A_{\alpha} - \frac{1}{2} \Big [ \Big (A_{\alpha}^{\dagger}A_{\alpha} \Big )^{*} \otimes \mathbb{I} + \mathbb{I} \otimes A_{\alpha}^{\dagger}A_{\alpha} \Big ] \Big )
 \end{equation}

Où $A_{\alpha}$ sont les opérateurs de Lindblad, tandis que $k_{\alpha}$ représente une certaine constante de taux du processus de relaxation du bain, qui prennent des formes fonctionnelles différentes selon le modèle de Lindblad utilisé.



* $L_{trap}$ qui  capture la dynamique du transfert irréversible de l'énergie excitonique vers le centre réactionnel, où elle est piégée et localisée sur les sites de séparation de charges.



* $L_{decay}$ qui  rend compte de la décroissance excitonique ; la recombinaison de l'électron et du trou d'électron entraînant la perte irréversible d'énergie dans l'environnement sous forme de chaleur.



* $L_{rad}$ = qui capture les processus d'absorption et d'emission (spontanée ou stimulée)

 \begin{equation}
     L_{rad} = L_{rad}^{abs} + L_{rad}^{emit}
 \end{equation}


# Influence de l'environnement

La complexité de la dynamique de relaxation non-unitaire du bain d'un système quantique ouvert comme le complexe FMO provient de son couplage avec le bain. Cela est modélisé comme un nombre infini d'oscillateurs harmoniques bosoniques qui peuvent se coupler aux transitions électroniques (excitons) du système. Afin de décrire l'évolution de la matrice de densité réduite du système sans définir explicitement les degrés de liberté du bain, l'influence du bain doit être prise en compte d'une manière différente.

## potentiel harmonique

Le principe de Franck-Condon des transitions électroniques verticales est supposé, de sorte que l'excitation d'un site BChl dans le système est instantanée et se produit sans changement de configuration du bain. Ce n'est qu'au bout d'un certain temps que le bain change de configuration (ou se réorganise). Le changement d'énergie associé à cette relaxation est connu sous le nom d'énergie de réorganisation λ

Cette influence du bain peut être complètement caractérisée par une fonction de corrélation temporelle C(t) qui decroit avec le temps.

En raison de sa forme fonctionnelle compliquée, et donc généralement difficile à manipuler, la fonction de corrélation temporelle est plus commodément exprimée dans le domaine des fréquences en prenant la transformée de Fourier-Laplace :

 \begin{equation}
     C(\omega) = \int_0^{\infty} C(t)e^{-i\omega t} dt
 \end{equation}

Toute fois, d'apres la formule d'euler, $Im(e^{-i\omega t})$ = $sin( \omega t)$ par consequent par resolution on obtiens deux termes soit un paire et un autre impaire.


la partie impaire prend la forme

 \begin{equation}
     C_{qq}^{''}(\omega) = \frac{\pi}{2} (\delta(\omega - \tilde{\omega})- (\delta(\omega + \tilde{\omega}))
 \end{equation}

 et la partie paire prend la forme

  \begin{equation}
     C_{qq}^{'}(\omega) = coth \Big (\frac{\beta \hbar \omega}{2} \Big ) C_{qq}^{''}(\omega)
 \end{equation}

 La fonction de corrélation totale peut donc être écrite uniquement en termes de la partie impaire indépendante de la température comme :


   \begin{equation}
     C_{qq}^(\omega) = \Big [coth \Big (\frac{\beta \hbar \omega}{2}\Big )+1 \Big ] C_{qq}^{''}(\omega)
   \end{equation}

### densite spectrale

Étant indépendante de la température et liée à la fonction de corrélation totale par une constante, la partie impaire de la fonction de corrélation, permet de récupérer toutes les informations sur le bain. Elle est donc utilisée pour définir la densité spectrale, J(ω), qui capture l'influence du bain harmonique sur le système. Évaluée à une fréquence de transition électronique donnée entre les excitons du système, ω, la densité spectrale est formée de la somme de tous les modes du bain, $\zeta$, avec une fréquence d'oscillation, $\omega_{\zeta}$, qui sont en résonance avec ω. La force de couplage entre le système et le bain est donnée par $g_{\zeta}$, et la fonction delta de Dirac fait en sorte que, pour une fréquence d'exciton donnée ω, seuls les modes de bain résonnants $(\omega_{\zeta}$ = ω) contribuent à la densité spectrale :


 \begin{equation}
     J(\omega) = \sum_{\zeta} g_{\zeta}^2 \delta(\omega - \omega_{\zeta})
 \end{equation}

 

Pour une fréquence d'exciton donnée ω, plus il y a de modes de bain résonnants et plus ils sont couplés fortement au système, plus l'intensité de la densité spectrale est grande. La force de couplage $g_{\zeta}$ et, par extension, la densité spectrale sont généralement approximées comme étant indépendantes du site dans le système i de sorte que $g_{\zeta}^{i}$ = $g_{\zeta}$ et on aura aussi  $J(\omega)_{i}$ = J(ω). <br>

La forme de la densité spectrale peut être approximée, la forme la plus courante étant la densité spectrale de Debye

\begin{equation}
     J_{Debye}(\omega) = 2\lambda \frac{\omega \gamma}{\omega^2 + \gamma^2}
\end{equation}

avec $\gamma$ le taux de relaxation du bain qui est fonction du temps de correlation de l'environnement par $\gamma = \frac{1}{\tau_c}$. Cette expression définit la fréquence du pic de la densité spectrale. on a donc deux cas

* pour $\omega$ > $\gamma$ : la densité spectrale décroît jusqu'à zéro
* pour $\omega$ < $\gamma$ : la densité spectrale est approximativement linéaire en $\omega$




## taux de transfert des populations

L'expression de la théorie de Redfield, $k(\omega_{ab})$, représente le transfert de population de l'exciton a vers l'exciton b et est définie par la fonction de corrélation à transformée de Fourier complète :


\begin{equation}
     k(\omega_{ab}) = C(\omega_{ab}) = 2[(1 + n_{BE}(\omega_{ab})J(\omega_{ab}) + n_{BE}(\omega_{ba})J(\omega_{ba})]
\end{equation}

avec $n_{BE}(\omega)$ la distribution de Bose-Einstein donné par

\begin{equation}
     n_{BE}(\omega) = \frac{1}{exp(\beta \omega) - 1}
\end{equation}


Cette expression représente le rapport de population sans dimension entre deux excitons séparés en énergie de $\omega$, où $\beta= \frac{1}{K_bT}$.

$n_{BE}(\omega_{ab})$ représente la population de ***b*** par rapport à ***a*** à une température T donnée. Si l'exciton ***a*** a une énergie plus élevée que l'exciton ***b*** de sorte que $\omega_{ab}$ > 0, lorsque l'écart d'énergie $\omega_{ab}$ tend vers zéro, ou lorsque la température tend vers l'infini, la population relative de l'exciton ***b*** par rapport à l'exciton ***a*** tend vers zéro ($n_{BE}(\omega_{ab})$ → 0) ; c'est-à-dire qu'ils ont la même probabilité d'être peuplés. À l'autre extrême, lorsque la séparation d'énergie des excitons tend vers l'infini, ou lorsque la température tend vers le zéro absolu, la distribution de Bose-Einstein tend vers l'infini ; l'exciton ***b*** a 100% de probabilité d'être peuplé par rapport à ***a***.

Alors le taux constant de Redfield $k(\omega_{ab})$ depend de la temperature et de la difference d'energie entre les excitons dans le treansfert de population capturé par $n_{BE}(\omega)$

Cependant, apres un certain temps les processus de relaxation conduisent le systeme dans un etat stationnaire.
La description de la dynamique du systeme par une equation maitresse notamment celle sous Born-Markov comme l'***equation maitresse de Lindblad***, le systeme en contact avec l'environnement constitué de plusieurs osscillateurs harmonique converge vers un etat stationnaire correspond a un etat canonique thermal 

\begin{equation}

# Equation Maitresse de Lindblad

L'équation de Lindblad, développée par G. Lindblad en 1976, et séparément par Gorini et Kossakowski la même année, est une théorie cohérente qui permet de modéliser la dynamique non unitaire des systèmes quantiques ouverts. Connue sous le nom de application linéaire complètement positive et préservant la trace, les propriétés de l'équation de Lindblad permettent l'évolution de la matrice de densité réduite du système dans le temps sans détruire son état, en veillant au respect des lois de la mécanique quantique tout au long de l'évolution. Son utilité réside dans la possibilité de modéliser la dynamique de relaxation influencée par le bain du système purement en termes des degrés de liberté du système. Les degrés de liberté du bain n'ont pas besoin d'être explicitement définis ; au lieu de cela, son influence est capturée par l'utilisation de la densité spectrale.

## forme generale

La forme generale de l'equation de Lindblad pour une dynamique de relaxation de l'environnement est donnée par

$$
\frac{\partial}{\partial t} \rho = (L_{sys} + L_{bath}) \rho
$$

Avec $L_{bath}$ le superoperateur de l'environnement contenant les processus de relaxation de l'environnement. La forme mathematique de cet superoperateur est une application linéaire complètement positive et préservant la trace. De plus, le nombre d'occupation d'excitons dans le systeme est conservé. Donc pour toute evolution temporelle, on a toujours $tr(\rho(t)) = 1$

## Approximations

Du faite de la complexité de l'etude d'un tel complexe (systeme + environnement), la description de la dynamique de l'EET dans le complexe FMO necessite certains hypotheses et approximations notamment:

### Approximation de Born

On suppose que à l'instant initiale, le systeme et l'environnement sont separables c'est a dire

$$
\rho(0) = \rho_S(0) \otimes \rho_E(0)
$$

De plus, on neglige l'evolution temporellede l'environnement devant celle du systeme puisque l'environnement interagit plus sur le systeme que le systeme sur l'environnement. a tout instant, le systeme et l'environnement sont separables

$$
    \rho(t) = \rho_S(t)\rho_E(0)
$$


### Approximation de Markov

cette approximation considere que les effets memoires sont nuls c'est a dire l'etat de systeme a un instant t ne depend pas de celui a l'instant precedent. Le taux de relaxation de l'environnement est negligeables devant la relaxation du systeme

$$
    {\rho}_S(s) \rightarrow {\rho}_S(t)
$$
Deplus, on a 
$$
    \tau_E \ll \tau_R, \tau_S
$$

### Approximation seculaire

Cette approximation neglige les termes provenant du produit d'operateurs du système qui tourne rapidemment.
Dans ce cas, c'est equivalent a l'approximation des ondes rotatoires et est donc valide, Le taux de relaxation doit être plus petit que la fréquence du système.



## Construction des Operateurs de Lindblad

La dynamique d'un système quantique ouvert, ne tenant compte que de la dynamique unitaire régie par le hamiltonien du système Hsys et de la dynamique de relaxation du bain, est donnée par l'équation de Liouville

$$
\frac{\partial}{\partial t} \rho = (L_{sys} + L_{bath}) \rho
$$

Avec $L_{bath}$ le superoperateur de Liouville modelisé par le superoperateur de Lindblad

  \begin{equation}
     L_{bath} =  \sum_{\alpha} k_{\alpha} \Big (A_{\alpha}^{*} \otimes A_{\alpha} - \frac{1}{2} \Big [ \Big (A_{\alpha}^{\dagger}A_{\alpha} \Big )^{*} \otimes \mathbb{I} + \mathbb{I} \otimes A_{\alpha}^{\dagger}A_{\alpha} \Big ] \Big )
 \end{equation}

Où $A_{\alpha}$ sont les opérateurs de Lindblad, tandis que $k_{\alpha}$ représente une certaine constante de taux du processus de relaxation du bain, qui prennent des formes fonctionnelles différentes selon le modèle de Lindblad utilisé.

On distingue generalement trois modeles differents de l'equation maitresse :


### ***Dephasage Local***












L'equation maoitresse de Lindblas basée sur le déphasage local prend en compte plusieurs approximations dont:

* Le superoperateur decrivant l'environnement est equiivalent a celui du dephasage $L_{bath}$ = $L_{deph}$, dans ce cas, les operateurs de Lindblad $A_{\alpha}$ sont evalués pour chaque site *i*
* Le taux de dephasage $ \tau_{deph}$ est independant du site ***i***
* Le terme ***local*** est utilisé car chaque site est couplé directement à son propre bain indépendant

### ***Thermalisation Globale***

L'equation maitresse de Lindblad basée sur la Thermalisation globale prend en commpte des considerations telles que:
* Le superoperateur $L_{bath}$ = $L_{global}$, dans ce cas, chaque opérateur de Lindblad est construit pour une paire d'états propres d'exciton distincts ***a*** et ***b*** du Hamiltonien du système.
* Le terme de constante de taux de Redfield $k(\omega_{ab})$ représente le taux de transfert de population entre les états propres ***a*** → ***b***
* $k(\omega_{ab})$ depend de la difference d'energie entre les etat propres et la force de couplage entre le systeme et l'environnement decrite par la densité spectrale $J(\omega_{ab}) mais independant du site.
* On parle de ***global*** car il considère que tous les sites interagissent avec un bain partagé (c'est-à-dire global)

### ***Termalisation Locale***

L'equation maitresse de Lindblad basée sur la Termalisation Locale prend en compte:

* Le superoperateur $L_{bath}$ = $L_{local}$ Les opérateurs de Lindblad sont construits pour chaque combinaison de site ***i***, et d'écart de fréquence unique entre les états propres du Hamiltonien du système $\omega$.
* Tous les transferts de population entre les états propres ***a*** et ***b*** dont les énergies sont séparées par l'écart de fréquence unique pour lequel l'opérateur de Lindblad est construit ($\omega_{ab}$ = $\omega_{a}$ - $\omega_{b}$) sont sommés
* Chaque transfert de population $|b\rangle \langle a|$ contribue a la construction de l'operateur de Lindblad pondérée par la contribution de $c_i^*(b)c_i(a)$
* $c_i(a)$ est le ***i-ème*** coefficient de site de l'exciton (état propre) ***a*** du Hamiltonien du système, tandis que $c_i^*(b)$ est le conjugué complexe du ***i-ème*** coefficient de site de l'exciton ***b***.
* La constante de taux $k(\omega)$ représente le taux de transfert de population entre les états propres séparés par la fréquence unique $\omega$, et comme la constante de taux dans le modèle de thermalisation global, elle est indépendante du site.

avec $p_k$ et $m_k$ l'impulsion et la masse du kieme oscillateur harmonique, $V(x)$ l'energie potentielle telle que $ V(x) = \frac{1}{2}m_k\omega^2_kx^2_k $, alors

$$

$$ b = \frac{1}{\sqrt{2}}(X-iP) $$
$$ b^{\dagger} = \frac{1}{\sqrt{2}}(X+iP) $$

 P = \frac{p}{\sqrt{m\omega \hbar}} 
\end{align}

$$H_E = \sum_k \omega_k b_k^{\dagger}b_k$$

### Operateur Statistique

L'opérateur statistique est une représentation mathématique permettant de décrire l'état d'un système quantique. Ses propriétées sont: 

-- pour un etat pur $\rho^2=\rho$

-- operateur hermitien $\rho^{\dagger} = \rho$.

-- trace egale a 1 (sommes des probabilité $\mathrm{Tr}\rho = 1$.

-- Valeure moyenne des operateurs du systeme peuvent etre definit par $\langle A \rangle =\mathrm{Tr}(A\rho)$.



Dans le cas particulier d'un systeme quantique ouvert, l'operateur statistique total (systeme + environnement) est donné par le produit tensoriel entre l'operateur statistique du systeme et celui de l'environnement.

$$\rho(t) = \rho_S(t) \otimes \rho_E(t)$$

L'interet etant d'etudier la dynamique du systeme et non du complexe globale (systeme + environnement) alors pour avoir les informations sur le sous espace du systeme, il est necessaire de faire une trace partielle sur l'environnement et ne garder que la partie du systeme. on a alors

$$\rho_S(t) = \mathrm{Tr}_E[\rho(t)]$$ 

avec $\rho_S(t)$ qui decrit toutes les informations accessibles liées au systeme.

# ***Caracterisation de chaque modele de Lindblad***

* ### ***Modele de Lindblad à Dephasage Local***

Lorsque $L_{bath}$ = $L_{deph}$, le système est traité comme étant à la limite de température infinie ($T → \infty$). Dans cette limite, l'information sur les énergies relatives des sites est détruite et, par conséquent, l'état stationnaire du système a tous les sites également peuplés. L'état stationnaire correspond donc à l'état de mélange maximal :

\begin{align}
\rho_{mm} = \begin{pmatrix} 1/2 & 0 \\ 0 & 1/2 \end{pmatrix}
\end{align}

* ### ***Modele de Lindblad à Thermalisation Locale***

C'est le comportement attendu.

La combinaison d'un équilibrage vers l'état thermique canonique "correct" et de la prise en compte des distances et des couplages entre sites fait du modèle local un juste milieu prometteur entre les modèles de déphasage et global.

# ***Complexe FMO***

Apres avoir caracteriser chaque modele, une comparaison avec le modele HEOM est necessaire pour l'evaluations de leur efficacité à decrire les échelles de temps d'équilibration et le transfert de cohérence des populations de sites dans le complexe FMO à différentes températures et dans différents régimes de non-markov.<br>
Les résultats provenant principalement du modèle de thermalisation locale de Lindblad seront comparés aux résultats obtenus par ***Zhu et al***. A l'aide d'une approche HEOM modifiée et mise à l'échelle (développée par ***Shi et al.*** ) pour simuler la dynamique de l'EET dans le complexe FMO.

En raison de l'orientation des sites de BChl 1 et 6 vers la protéine de plaque basale du chlorosome initialement excité par la lumière du soleil, il est courant d'observer la dynamique du EET dans le FMO à partir d'états initiaux excités sur le site 1, le site 6 ou une superposition des deux. En tant qu'états purs, ceux-ci peuvent être exprimés sous forme de fonctions d'onde dont les formes respectives sont $|\psi(0)\rangle$ = $|1\rangle$, $|\psi(0)\rangle$ = $|6\rangle$ et $|\psi(0)\rangle$ = $\frac{1}{\sqrt{2}}(|1\rangle + |6\rangle)$. Lorsque l'operateur statistique pour l'état de superposition initial est formée (à partir du produit extérieur de la fonction d'onde $|\psi_{1,6}(0)\rangle$, des cohérences dans la base des sites seront présentes.

Afin de maintenir la cohérence avec d'autres études théoriques, le Hamiltonien du système FMO du tableau suivant et une densité spectrale de Debye comme montrer recemment seront utilisés. Une valeur de $\lambda = 35 cm^{-1} (= 6,593 rad ps^{−1})$ sera utilisée pour l'énergie de réorganisation, avec une valeur de $\gamma^{-1} = 50 fs rad^{−1} (= 20 rad ps^{−1})$ pour le taux de relaxation du bain. Ces paramètres assurent un accord entre l'ajustement numérique et les données expérimentales dans l'étude de ***Read et al***. Sur la visualisation de la structure excitonique du FMO. D'autres taux de relaxation du bain,    $\gamma^{-1} = 100 fs rad−1 (= 10 rad ps^{−1})$ et 166 fs $rad^{−1} (= 6,024 rad ps^{−1})$, seront utilisés plus tard pour simuler différents régimes de non-markoviens, conformément aux travaux de ***Zhu et al***. Bien qu'un taux de déphasage n'ait pas besoin d'être défini pour les approches HEOM, une valeur de $\tau_{deph} = 11,0 rad ps^{−1}$ a été utilisée comme paramètre pour les modèles de Lindblad. Ce taux de déphasage est équivalent à la constante de vitesse dans le modèle de déphasage local de Lindblad, et est également utilisé pour définir le taux de transfert de population entre les états propres dégénérés dans l'expression de la théorie de Redfield, c'est-à-dire $k(\omega = 0) = \tau_{deph}$. La valeur de $\tau_{deph} = 11,0 rad ps^{−1}$ a été calculée expérimentalement pour le système de collecte de lumière bactérien 2 (LH2) 67,68 et a donné un comportement raisonnable lorsqu'elle a été utilisée dans les simulations FMO. Cela est utilisé à la place d'une expression analytique pour le taux de déphasage :

$$
\tau_{deph} = \lim_{\omega = 0^+}k(\omega) = \frac{4\lambda K_BT}{\hbar \gamma}
$$

Toute fois, La délocalisation des excitons peut réduire le taux de déphasage, et les effets non markoviens peuvent également le modifier, donc on utilise à la place la valeur expérimentale.

# ***Analyses et interpretations***

Ce tracé compare l'évolution à long terme des populations des sites du FMO pour les trois états initiaux à la température physiologique (***300 K***), de chaque modèle de Lindblad.

* ***Le modèle de thermalisation globale*** (colonne de droite) atteint l'état thermique canonique beaucoup trop rapidement, en raison de sa ***négligence des distances inter-sites et des couplages en ne considérant que la différence d'énergie entre les excitons*** Cela peut également être vu sur la Figure qui presente le rapport de participation inverse qui, en suivant les rapports de participation inverse de chaque état au fil du temps, montre la délocalisation rapide à court terme de l'énergie des excitons sur plus de 2 sites. Cela est comparé aux modèles locaux, qui restent plus localisés. Une telle équilibration rapide signifie ***qu'aucun battement quantique n'est observé***. Le battement quantique, signature du transfert de cohérence quantique, est défini comme la modulation périodique des populations de sites et des amplitudes de crête.

* ***le modèle de déphasage local*** (colonne du milieu) s'équilibre sur une échelle de temps plus appropriée. Avec le taux de déphasage expérimental de $\tau_{deph} = 11,0 rad ps^{−1}$ utilisé, les populations des sites se déphasent sur une échelle de temps comparable à l'approche HEOM, de sorte que le battement quantique peut être observé. Malgré cela, le comportement d'équilibre incorrect est affiché. Modélisé dans la limite de température infinie, toutes les populations de sites sont également peuplées (avec une probabilité de 1/7) dans l'état stationnaire du système. ***Ce modèle ne parvient donc pas à capturer la population dominante des sites 3 et 4 à l'équilibre, comme cela est nécessaire pour un transfert efficace de l'énergie excitonique vers le centre réactionnel***.



On remarque alors que les modèles de déphasage local et de thermalisation globale ont des forces et des faiblesses complémentaires ; là où le modèle global s'équilibre à l'état thermique canonique mais le fait sur des échelles de temps qui ne tiennent pas compte correctement des effets de cohérence quantique, le modèle de déphasage décrit ces effets sur des échelles de temps courtes, mais s'équilibre à l'état stationnaire incorrect. À première vue, ***le modèle de thermalisation locale de Lindblad semble capturer les forces des deux modèles de déphasage et global, s'accordant qualitativement bien avec l'approche HEOM***. Il s'équilibre sur une échelle de temps appropriée et à l'état canonique, avec un taux de déphasage qui ne détruit pas trop rapidement le battement quantique.

* ***État initial de site 1***<br>
 Étant très fortement couplé (V12 = -87,7 cm−1, le couplage inter-sites le plus fort dans FMO), le battement quantique entre les sites 1 et 2 persiste pendant au moins 800 fs dans les approches HEOM et Lindblad. Il y a un chevauchement plus important des populations de sites à ~40 fs dans l'approche HEOM en raison d'un couplage électronique (c'est-à-dire inter-sites) effectif plus élevé dû à l'interaction avec la densité spectrale (vu precedemment). La haute énergie du site 2 par rapport au site 1 entraîne une longue durée de vie de la population du site 1 et donc une lente accumulation de la population sur le site 3. De plus, la très grande différence entre les énergies des sites 2 et 3 $(|ΔE_{23}| = 320 cm^{−1})$ signifie que tout transfert de population du site 1 vers le site 2 entraîne un transfert presque irréversible vers le site 3, le transfert inverse 3 → 2 étant supprimé. Par conséquent, il n'y a pas d'accumulation de population sur le site 2, mais il y en a sur le site 3. Tous les autres sites (4-7), pour les deux modèles, ne connaissent pas de changement notable de population, reflétant la proximité spatiale et la force du couplage d'excitation entre les sites impliqués dans la voie de transfert EET (1 → 2 → 3). La raison pour laquelle il n'y a pas de battement de population entre les sites 1 et 6, malgré leur proximité spatiale, est que l'écart d'énergie entre eux est supérieur à celui entre les sites 1 et 2 $(|ΔE_{16}| = 220$ contre $|ΔE_{12}| = 120 cm^{−1})$, et le couplage plus faible $(V_{16} = -13,7 cm^{−1})$.

* ***Etat Initial de site 6*** <br>
Les battements quantique des populations entre la force de couplage des sites 5 et 6 $(V_{56} = 81.1cm^{-1})$ peuvent etre observés, Les deux modèles indiquent une cohérence quantique qui persiste pendant au moins 650 fs, en accord avec les résultats expérimentaux. Cependant, les battements décrits par l'approche HEOM (Hierarchical Equations of Motion) entraînent un recouvrement plus important des populations des différents sites.
Cela s'explique à nouveau par un couplage électronique plus fort entre les sites dans le modèle HEOM.
On observe également une faible quantité de battements quantiques entre les sites 3 et 4, qui sont fortement couplés et ont des énergies similaires ($ V_{34} = -53.5cm^{-1}$, $|ΔE_{34}| = 110 cm^{−1})$.<br>

# ***CONCLUSION***

L'équation Maitresse de Lindblad fournit une description peu coûteuse en calcul, traitable et facile à mettre en œuvre de la dynamique des systèmes quantiques ouverts. Un modèle de ce type reproduit avec précision la dynamique de transfert d'énergie d'excitation (EET) dans le complexe FMO, telle que décrite par l'approche HEOM.

Cependant, pour le modèle du système à 2 sites, l'incapacité de l'équation de Lindblad à décrire le comportement non markovien s'est avérée importante pour certains jeux de paramètres. Des écarts plus importants de l'état stationnaire par rapport à l'état thermique canonique décrit par le HEOM se produisent dans les régimes de couplage plus forts et à plus basse température.

Le modèle de déphasage local, paramétré uniquement par un taux de déphasage et avec un choix de valeur appropriée déterminée expérimentalement, peut décrire les transferts de cohérence quantique à court terme avec un accord raisonnable avec HEOM, à la fois dans le modèle à 2 sites et dans le complexe FMO à 7 sites. Cependant, l'équilibrage vers l'état maximalement mélangé invalide le comportement à long terme de l'état quantique évolué par ce modèle.

Le modèle de thermalisation global, bien qu'équilibrant vers l'état thermique canonique (montré par Zhu et al. comme étant l'état d'équilibre pour le complexe FMO défini dans les paramètres vu plus haut), néglige les distances et les couplages inter-sites dans sa prise en compte des transferts de population. Cela se traduit par des temps d'équilibrage très rapides dans le complexe FMO, le modèle n'étant pas en mesure de prédire les effets de cohérence quantique.

Le modèle de thermalisation local a recréé de nombreux comportements de cohérence dans le complexe FMO, prédits par l'approche HEOM, à la fois à des températures cryogéniques et physiologiques. En traitant la force de couplage comme un paramètre libre, le modèle de thermalisation local a permis d'ajuster les populations des sites et a montré qu'il y avait seulement de petites différences avec l'approche HEOM dues aux effets non markoviens. Appliqué au complexe FMO, ce modèle de Lindblad a très bien décrit la dynamique physiologique dans un régime fortement non markovien, en bon accord avec HEOM.

De plus, une simulation de 2,5 ps du complexe FMO dans des conditions physiologiques peut être effectuée en moins d'une seconde, contre 1,5 heure avec l'approche HEOM.

Bien que l'approche HEOM soit un traitement exact et non perturbatif capable de fournir une description précise de la dynamique EET dans les complexes photosynthétiques comme le FMO, tout en prenant en compte les effets non markoviens et d'équilibre non canonique, la capacité de l'équation de Lindblad (ou plus spécifiquement du modèle de thermalisation local) à récréer un tel comportement avec une impressionnante rapidité et en bon accord avec HEOM est prometteuse. Cette approche pourrait être facilement mise à l'échelle pour des systèmes plus importants, comme les complexes collecteurs de lumière des bactéries pourpres, et être combinée avec d'autres processus dynamiques pour calculer les efficacités de transfert d'énergie d'excitation. De plus, l'approche de Lindblad pourrait être utilisée pour simuler rapidement et précisément la dynamique EET dans le cadre de la conception in silico de nouvelles architectures de captage de la lumière.
